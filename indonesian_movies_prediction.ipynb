```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_movie, test_size=0.2)
# train_stats = train.describe()


train_Y = format_output(train)
test_Y = format_output(test)
```

```python
import tensorflow as tf
from keras.utils.vis_utils import plot_model
```

```python
from keras import backend as K

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation = "relu", input_shape = (1056,)),
    tf.keras.layers.Dense(64, activation = "relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(32, activation = "relu"),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(3, activation = "softmax"),
])


optimizer = tf.keras.optimizers.Adam(learning_rate= 1e-3)
model.compile(loss = 'sparse_categorical_crossentropy', optimizer= optimizer,  metrics=['acc',f1_m,precision_m, recall_m])

model.summary()
```

```python
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model.fit(train, train_Y,
                    epochs=50, batch_size=10, validation_data=(test, test_Y))
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=717, stop_words="english")
```

```python
vector = cv.fit_transform(movies['depend'].values.astype('U')).toarray()
```

```python
mean = np.mean(vector)
vector = vector - mean
vector.shape
```

```python
split_size = int(len(vector) * 0.8)
print(split_size)
train = vector[:split_size]
valid = vector[split_size:]
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(512, activation = "relu", input_shape = (None,717)),
    tf.keras.layers.Dense(256, activation = "relu"),
    tf.keras.layers.Dense(128, activation = "relu"),
    tf.keras.layers.Dense(64, activation = "relu"),
    tf.keras.layers.Dense(3, activation = "softmax"),
])


optimizer = tf.keras.optimizers.Nadam(learning_rate= 1e-3)
model.compile(loss = 'sparse_categorical_crossentropy', optimizer= optimizer, metrics= ['acc'])
```

```python
model.fit(train, train_Y, epochs = 20, validation_data=(valid, test_Y) )
```
